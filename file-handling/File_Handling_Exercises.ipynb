# File Handling - Exercises

## Problem 1

You've decided to keep a journal. You have your first three entries ready to save:

```python
entries = [
    "Day 1: Started learning Python",
    "Day 2: Went for a run",
    "Day 3: Built my first Python script"
]
```

Write the entries to a file called `journal.txt`, one per line. Then read the file back and print its contents.

**Sample output:**
```
Day 1: Started learning Python
Day 2: Went for a run
Day 3: Built my first Python script
```

````python
entries = [
    "Day 1: Started learning Python",
    "Day 2: Went for a run",
    "Day 3: Built my first Python script"
]

# TODO: Write the entries to journal.txt, one per line, then read and print the contents
file = open("journal.txt", "w")
file.write("\n".join(entries))
file.close()

file = open("journal.txt", "r")
contents = file.read()
file.close()
print(contents)

````

## Problem 2

A few more days have passed and you have new entries to add to your journal:

```python
new_entries = [
    "Day 4: Had coffee with a friend",
    "Day 5: Completed a Python project",
    "Day 6: Read a book"
]
```

Append these entries to `journal.txt`, one per line. Then read the file back and print all its contents to confirm the original entries are still there.

**Sample output:**
```
Day 1: Started learning Python
Day 2: Went for a run
Day 3: Built my first Python script
Day 4: Had coffee with a friend
Day 5: Completed a Python project
Day 6: Read a book
```

````python
new_entries = [
    "Day 4: Had coffee with a friend",
    "Day 5: Completed a Python project",
    "Day 6: Read a book"
]

# TODO: Append the entries to journal.txt, then read and print all contents
file = open("journal.txt", "a")
file.write("\n" + "\n".join(new_entries))
file.close()

file = open("journal.txt", "r")
contents = file.read()
file.close()
print(contents)
````

## Problem 3

You want to look back at your Python-related entries. Using context managers (`with` statements), read `journal.txt` and create a new file called `python_entries.txt` containing only the entries that mention `"Python"`. Then read the new file back and print its contents.

````python
# TODO: Use context managers to read journal.txt, filter to entries mentioning "Python", write to python_entries.txt, then read and print it
with open("journal.txt", "r") as file:
    contents = file.read().split("\n")

python_entries = [entry for entry in contents if "Python" in entry]

with open("python_entries.txt", "w") as file:
    file.write("\n".join(python_entries))
````
